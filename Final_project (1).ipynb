# Data Preparation

```python
import pandas as pd
import yfinance as yf

# Fetch S&P 500 ticker symbols from Wikipedia
url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
sp500 = pd.read_html(url)[0]
tickers = sp500['Symbol'].tolist()

# Define date range
start_date = '2010-01-01'
end_date = '2023-12-31'

# Fetch data for all tickers
data = [yf.download(ticker, start=start_date, end=end_date, progress=False)
        .assign(Symbol=ticker) for ticker in tickers]

# Concatenate data into a single DataFrame
all_data = pd.concat(data)

# Filter data for the specified date range
all_data = all_data[(all_data.index >= start_date) & (all_data.index <= end_date)]

# Reset index to make 'Date' a column again
all_data.reset_index(inplace=True)

# Display the DataFrame
all_data
```

```python
all_data['Date'] = pd.to_datetime(all_data['Date'])
all_data['year'] = all_data['Date'].dt.year
all_data['month'] = all_data['Date'].dt.month
all_data['day'] = all_data['Date'].dt.day
all_data
```

```python
all_data.sort_values(by=['Symbol','year','month','day'], inplace=True)
all_data = all_data.reset_index(drop=True)
all_data
```

```python
my_df = all_data[ (all_data['year']<=2023) & (all_data['year']>=2010) ].copy()
```

```python
my_df.reset_index(drop=True, inplace=True)
my_df
```

```python
my_df['prev_price']=my_df.groupby('Symbol')['Adj Close'].shift(1)
my_df['daily_return']= (my_df['Adj Close'] - my_df['prev_price']) / my_df['prev_price']
my_df['d_vol']= my_df['Volume'] * my_df['Close']
```

```python
my_df['avrg_return_daily'] = my_df.groupby(['Symbol','year','month'])['daily_return'].transform('mean')
my_df['std_return_daily'] = my_df.groupby(['Symbol','year','month'])['daily_return'].transform('std')
my_df['skew_return_daily'] = my_df.groupby(['Symbol','year','month'])['daily_return'].transform('skew')
my_df['avrg_vol'] = my_df.groupby(['Symbol','year','month'])['d_vol'].transform('mean')
my_df['std_vol'] = my_df.groupby(['Symbol','year','month'])['d_vol'].transform('std')
my_df
```

```python
monthly_df=(my_df.groupby(['Symbol','year','month']).last().reset_index()).drop(columns=['daily_return','Open','High','Low','Close','Volume','day','prev_price','d_vol'])
monthly_df
```

```python
monthly_df['prev_price']=monthly_df.groupby('Symbol')['Adj Close'].shift(1)
monthly_df['return']=( monthly_df['Adj Close'] - monthly_df['prev_price'] )/monthly_df['prev_price']
monthly_df['next_month_return'] = monthly_df.groupby('Symbol')['return'].shift(-1)
monthly_df
```

```python
import numpy as np
monthly_df['cum_3m']=monthly_df.groupby('Symbol')['return'].rolling(3).apply(lambda x: np.prod(1+x)-1).reset_index(drop=True)
monthly_df['cum_6m']=monthly_df.groupby('Symbol')['return'].rolling(6).apply(lambda x: np.prod(1+x)-1).reset_index(drop=True)
monthly_df['cum_9m']=monthly_df.groupby('Symbol')['return'].rolling(9).apply(lambda x: np.prod(1+x)-1).reset_index(drop=True)
monthly_df['cum_12m']=monthly_df.groupby('Symbol')['return'].rolling(12).apply(lambda x: np.prod(1+x)-1).reset_index(drop=True)
monthly_df
```

```python
monthly_df['sharpe_ratio']=( monthly_df['std_return_daily'] / monthly_df['avrg_return_daily'] )
monthly_df
```

```python
from itertools import combinations_with_replacement

# Function to create new features by interacting all combinations of two features and calculating squared values
def create_new_features(X):
    num_samples, num_features = X.shape
    new_features = []

    # Calculate squared values of each original feature
    for i in range(num_features):
        new_features.append(X[:, i] ** 2)

    # Interact all combinations of two features and calculate their product
    for feat1, feat2 in combinations_with_replacement(range(num_features), 2):
        new_features.append(X[:, feat1] * X[:, feat2])

    # Stack the original features and new features horizontally
    return np.column_stack([X] + new_features)

# Create new features for training and test data
X_train_new = create_new_features(X_train_normalized)
X_test_new = create_new_features(X_test_normalized)
```

```python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.model_selection import train_test_split

# Feature Engineering: Create new features
X_fl = final_data_filled.drop(columns=['Symbol', 'Date', 'next_month_return', 'return', 'year', 'month', 'Ticker'])
y_fl = final_data_filled['next_month_return']

poly_fl = PolynomialFeatures(degree=2, include_bias=False)
X_poly_fl = poly_fl.fit_transform(X_fl)
X_with_interaction_fl = pd.DataFrame(X_poly_fl, columns=poly_fl.get_feature_names(input_features=X_fl.columns))

# Normalization: Normalize the features
scaler_fl = StandardScaler()
X_with_interaction_normalized_fl = scaler_fl.fit_transform(X_with_interaction_fl)

# Splitting: Split the normalized feature matrix and target variable into training and testing sets
X_train_fl, X_test_fl, y_train_fl, y_test_fl = train_test_split(X_with_interaction_normalized_fl, y_fl, test_size=0.2, random_state=8)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_squared_error, r2_score
from scipy.stats import randint

# Define the model
random_forest = RandomForestRegressor(random_state=8)

# Hyperparameter tuning
param_dist = {
    'n_estimators': randint(10, 50),          # Reduced range for n_estimators
    'max_depth': [None, 5],                    # Fewer options for max_depth
    'min_samples_split': randint(2, 5),        # Reduced range for min_samples_split
    'min_samples_leaf': [1],
    'max_features': ['auto', 'sqrt']   # Only one option for min_samples_leaf
}

random_search = RandomizedSearchCV(random_forest, param_dist, n_iter=3, cv=5, scoring='r2')
random_search.fit(X_train_fl, y_train_fl)

# Best hyperparameters
best_params = random_search.best_params_
print("Best Hyperparameters:", best_params)

# Best model
best_random_forest = random_search.best_estimator_

# Train the model
best_random_forest.fit(X_train_fl, y_train_fl)

# Evaluate on training data
y_train_pred = best_random_forest.predict(X_train_fl)
train_rmse = mean_squared_error(y_train_fl, y_train_pred, squared=False)
train_r2 = r2_score(y_train_fl, y_train_pred)
print("Train RMSE:", train_rmse)
print("Train R^2:", train_r2)

# Evaluate on test data
y_test_pred = best_random_forest.predict(X_test_fl)
test_rmse = mean_squared_error(y_test_fl, y_test_pred, squared=False)
test_r2 = r2_score(y_test_fl, y_test_pred)
print("Test RMSE:", test_rmse)
print("Test R^2:", test_r2)
```
